# Finn Demo

In this notebook we will explore pathways on how to automate content production for Finn using Generative AI.

Status quo:

- Finn has currently high content production costs, as content is created externally.
- To reduce costs, parts of the content production process could be automated.

What can be automated:

- Similar repetitive content, e.g. certain blog posts, can be automated.
- with a better pathway even more content can be automated, down the line

Good to know:

- google values high quality content - this is independent of the content being generated by AI or not
- AI is GOOD at creating informational content, which is the type of content that Finn is aiming to produce in this case
